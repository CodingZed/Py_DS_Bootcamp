** How many people have the word Chief in their job title? (This is pretty tricky) **

```python
sal.head()
```

```python
def chief_string(title):
    if 'chief' in title.lower().split():
        return True
    else:
        return False
```

```python
sum(sal.JobTitle.apply(lambda x: chief_string(x)))
```

```python
sal_627=sal[sal.JobTitle.isin([x for x in sal.JobTitle if 'chief' in x.lower()])]
```

```python
sal_477=sal[sal.JobTitle.isin([x for x in sal.JobTitle if 'chief' in x.lower().split()])]
```

```python
sal_627[~sal_627.JobTitle.isin([x for x in sal_477.JobTitle if 'chief' in x.lower().split()])]
```

```python
sal_627[sal_627.EmployeeName=='ALSON LEE']
```

```python
def chief_count(title):
    return title.lower().split().count('chief')
    
```

```python
sum(sal.JobTitle.apply(lambda x: chief_count(x)))
```
